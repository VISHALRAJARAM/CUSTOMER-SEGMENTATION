```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
cust_data = pd.read_csv('/content/Mall_Customers.csv')
cust_data.head()
```

```python
#finding no of rows and cols
cust_data.shape
```

```python
#getting info abt dataset
cust_data.info()
```

```python
#checking for missing values
cust_data.isnull().sum()
```

```python
# classification based on annual income and spending score
x=cust_data.iloc[:,[3,4]].values #iloc used to locate the col .
print(x)
```

```python
#choosing the no of clusters - optimum no of clkuster is found with WCSS method
#finding WCSS value for different clusters

wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)

```

```python
#plot an elbow graph

sns.set()
plt.plot(range(1,11),wcss)
plt.title('The Elbow Point Graph')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show
```

```python
#optimum no of clusters will be 5 .

#k=5
kmeans = KMeans(n_clusters=5,init='k-means++',random_state=42)
y = kmeans.fit_predict(x)

print(y)
```

```python
#visualising all the clusters

plt.figure(figsize=(8,8))
plt.scatter(x[y==0,0],x[y==0,1],s=50,c='green',label='Cluster1')
plt.scatter(x[y==1,0],x[y==1,1],s=50,c='red',label='Cluster2')
plt.scatter(x[y==2,0],x[y==2,1],s=50,c='yellow',label='Cluster3')
plt.scatter(x[y==3,0],x[y==3,1],s=50,c='blue',label='Cluster4')
plt.scatter(x[y==4,0],x[y==4,1],s=50,c='magenta',label='Cluster5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='black',label='Centroids')
plt.title('Clusters of Customers')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
